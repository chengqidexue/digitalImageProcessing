## 1.4 图像处理的基本步骤
数字图像处理的基本步骤.png



1.   图像获取(采集)

2.   图像增强
    
    显现那些被模糊了的细节，或者简单的凸显一幅图像中感兴趣的特征。

3.   图像复原

    - 改变图像的外貌
    - 以图像退化的数学或者概率模型为基础
    - 相比图像增强，是一个客观过程

4.   彩色图像处理

5.   小波变换
    - 在各种分辨率小描述图像的基础
    - 用于图像压缩及金字塔描述方法
6.   图像压缩
    - 减少图像的贮存量
    - 在传输图象时降低频带
    - JPG图像压缩标注
7.   形态学处理
    - 涉及提取图像元素的工具
    - 在表现和描述形状方面非常有用
8.   图像分割
    - 将图像划分为组成部分或目标物
9.   表述和描述
    
    表示：确定数据应该由边界还是区域来表示。
    - 边界：外部形状特性，如拐角、曲线。
    - 区域：内部特性，如纹理和骨骼形状。
10.   识别

    基于目标描述给目标赋以符号的过程
